```python
#Dense Embedding
from sentence_transformers import SentenceTransformer
import numpy as np

def get_dense_embedding(text: str, model_name: str = "all-MiniLM-L6-v2"):
    # load model
    model = SentenceTransformer(model_name)
    # encode returns fixed-size dense vector
    embedding = model.encode(text)
    return embedding

if __name__ == "__main__":
    sample_text = "This is a test sentence for embedding."
    emb = get_dense_embedding(sample_text)
    print("Embedding vector:", emb)
    print("Length of vector:", len(emb))
    # for multiple sentences
    texts = ["Hello world!", "Embedding with sentence transformers."]
    embs = get_dense_embedding(texts)
    print("Embeddings shape:", np.array(embs).shape)  # e.g. (2, 384) for this model
```
